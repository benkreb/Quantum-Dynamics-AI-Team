```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
monkey_image = cv2.imread("/content/shocked_monkey.jpg")
plt.imshow(monkey_image)
```

```python
monkey_rgb = cv2.cvtColor(monkey_image , cv2.COLOR_BGR2RGB)
plt.imshow(monkey_rgb)
```

```python
monkey_image[:,:,0] = 0
monkey_image[:,:,1] = 0
plt.imshow(monkey_image)
```

```python
shocked_guy = cv2.imread("/content/shocked_guy.png")
shocked_guy = cv2.cvtColor(shocked_guy, cv2.COLOR_BGR2RGB)
plt.imshow(shocked_guy)
```

```python
flipped = cv2.flip(shocked_guy , -1)
plt.imshow(flipped)
```

```python
c_shocked = shocked_guy
rectangle =  cv2.rectangle(c_shocked,(500,50),(1050,900),color=(255,0,0),thickness=20)
plt.imshow(c_shocked)
```

```python
shockedcat_image = cv2.imread("/content/shocked_cat.jpg",0)
plt.imshow(shockedcat_image,cmap = "gray")
```

```python
ret,tresh1 = cv2.threshold(shockedcat_image,127,255,cv2.THRESH_BINARY)
plt.imshow(tresh1,cmap='gray')
```

```python
sobelCat = cv2.Sobel(shockedcat_image,cv2.CV_64F,1,0,ksize = 5)
plt.imshow(sobelCat ,cmap='gray')
```
