8) Open the "shocked_cat" image and display it in the notebook as black and white


```python
img = cv2.imread('DATA/shocked_cat.jpg',0)
plt.imshow(img, cmap = 'gray')
```

9)Apply a binary 'gray' threshold onto the image.


```python
ret,th1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
plt.imshow(th1,cmap='gray')
```

10) Create a Horizontal Sobel Filter with a kernel size of 5 to the grayscale version of the cat image and then display the resulting gradient filtered version of the image.

```python
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize = 5)
plt.imshow(sobelx ,cmap='gray')
```
