```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
monkey_image = cv2.imread("C:/Users/iremb/Desktop/AIteam/shocked_monkey.jpg")
plt.imshow(monkey_image)
```

```python
monkey_rgb = cv2.cvtColor(monkey_image, cv2.COLOR_BGR2RGB)
plt.imshow(monkey_rgb)
```

```python
monkey_arr = np.asarray(monkey_rgb)
blue_monkey = monkey_rgb.copy()
blue_monkey[:, :, 0]=0
blue_monkey[:, :, 1]=0 
plt.imshow(blue_monkey)
```

```python
shocked_guy = cv2.imread("C:/Users/iremb/Desktop/AIteam/shocked_guy.png")
guy_rgb = cv2.cvtColor(shocked_guy, cv2.COLOR_BGR2RGB)
plt.imshow(guy_rgb)
```

```python
flipped_img = guy_rgb.copy()
flipped_img =  cv2.flip(flipped_img,0)
plt.imshow(flipped_img)
```

```python
guy_rgb = cv2.cvtColor(shocked_guy, cv2.COLOR_BGR2RGB)
new_guy_img = guy_rgb.copy()
pt1 = (500,190)
pt2 = (1050,950)
cv2.rectangle(new_guy_img, pt1=pt1, pt2=pt2, color=(255,0,0),thickness=10)
plt.imshow(new_guy_img)
```

```python
shocked_cat = cv2.imread("C:/Users/iremb/Desktop/AIteam/shocked_cat.jpg")
cat_rgb = cv2.cvtColor(shocked_cat, cv2.COLOR_BGR2GRAY)
plt.imshow(cat_rgb, cmap='gray')
```

```python
ret,thresh1 = cv2.threshold(cat_rgb,127,255,cv2.THRESH_BINARY)
plt.imshow(thresh1, cmap='gray')
```
